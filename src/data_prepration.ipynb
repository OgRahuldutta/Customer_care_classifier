```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler, OrdinalEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer

num_transformer = StandardScaler()
cat1_transformer = OneHotEncoder(handle_unknown='ignore')
cat2_transformer = OrdinalEncoder()

preprocessor = ColumnTransformer(
    transformers=[
        ('onehot', cat1_transformer, ['Internet_Service', 'Contract_Type', 'Payment_Method']),
        ('ordinal', cat2_transformer, ['Gender', 'Tech_Support']),
        ('num', num_transformer, ['Age', 'Tenure_Months', 'Monthly_Charges'])
    ]
)
```

```python
le = LabelEncoder()
df['Churn'] = le.fit_transform(df['Churn'])
```

```python
X = df.drop(columns=['Churn'])
y = df['Churn']

X = preprocessor.fit_transform(X)
```
